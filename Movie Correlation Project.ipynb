```python
#Importing libraries 
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusting the configuration of plots which will be created 

#Reading in the data
df = pd.read_csv(r'C:\Users\ASUS\Desktop\Data Analyst Portfolio\Movie Project\movies.csv')
```

```python
#Looking at the data
df.head()
```

```python
#Checking for missing data 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Analyzing the data type of columns 
df.dtypes
```

```python
#change data type of budget and gross income column
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
#Creating correct year column 

df['yearcorrect'] = df['released'].astype(str).str[:4];
```

```python
df
```

```python
#Sorting movies in descending order by revenue
df.sort_values(by=['gross'],inplace = False, ascending = False)
```

```python
#Command to show all rows in data
#pd.set_option('display.max_rows', None)
```

```python
#Dropping duplicate rows
df.drop_duplicates().head()
```

```python
#Hypotheses 
#1. Budget and Revenue are correlated 
#2. Company and revenue are correlated 
```

```python
#Scatter plot between Burget and Gross Revenue
plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earning')


plt.show()
```

```python
df.head()
```

```python
#Plotting a budget vs gross earning regression using seaborn

sns.regplot(x = 'budget', y= 'gross', data= df, scatter_kws = {"color":"red"}, line_kws = {"color":"blue"})
```

```python
#Quantifying correlation using in-built function (only for numerical fields)
df.corr(method= 'pearson') #By default we use pearson. Can also have kendall, spearman
```

```python
#High correlation between budget and gross
```

```python
correlation_matrix= df.corr(method= 'pearson')

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation matrix for numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
#Converting all fields into numeric values to find correlation 

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
correlation_matrix= df_numerized.corr(method= 'pearson')

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation matrix for all features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(ascending = False)
sorted_pairs
```

```python
high_correlation = sorted_pairs[(sorted_pairs)>0.5]
high_correlation
```
